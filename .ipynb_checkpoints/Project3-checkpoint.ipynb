## Introduction

"Sharing-economy" is a newly developed business model, in which the same product can be sold multiple times to generate more profits (Gansky, 2010). Within this model, consumers have more choices and flexibility, and suppliers gain more benefits than traditional one-way business model. "Sharing bikes" is one famous example from transportation industry in urban economy. Airbnb is also one of the incumbent business in the collaborative market to provide temporary travel accommodations to tourists (Zervas et al., 2016, Wang & Nicolau, 2017). As a successful pioneer under this business model, Airbnb communities were established across 34000 cities in 191 countries in 2015; serving more than 50 million guests (*Airbnb Summer Travel Report: 2015*). Airbnb provides guests and hosts an awesome platform to customize personalised travel experience around the world. The website generates huge amount of data and data networks everyday from its customers and the associated data analysis can help the company understand the behaviour of its customers better and make relevant business decisions. 

Airbnb experienced an exponential growth on its market size in recent years, researchers as well as economists conducted countless studies on the determinants of price of Airbnb listings from both demand and supply side (Wang & Nicolau, 2017).

We also take interests in factors affecting the price of Airbnb listings and we would like to focus on Airbnb community in NYC. New York City is one of the most popular and attractive city in the world, it generates huge amount of tourism spending in different industries.

By taking literature reviews, we can observe many price determinants from different perspectives. Some of the variables exist in our dataset while most of them does not. With this research paper, we can not only further conclude the effect of the variables that were analyzed on price with a new dataset, but also exploring the impact of new variables without many supports from previous literatures. 

In this project, we are going to use the `AB_NYC_2019.csv` contained in `New York City Airbnb Open Data` from Kaggle to be the main dataset for analysis. It comes from the official website of Airbnb and contains numerous listings and their associated features (2019 version). Since our __main question__ is "What leads to the difference of Airbnb listings' prices in New York City?", the variable __price__ in this dataset is our Y variable (outcome). Here are some of our important explanatory variables (X variables): __number of reviews__, __room type__, __neighbourhood__, __neighbourhood group__ and __annual availability__. We will visualize relationships between these variables to explore how each X variable affect prices of listings, as well as how X variables are correlated through appropriate __plots__. Additionally we will create some __maps__ to convey geographical information for Airbnb listings and find out the locational impact on listings' prices. 

Additionally, we will merge a new dataset `nyc-rolling-sales.csv` from `NYC Property Sales` to the original dataset. This new dataset records properties sold in New york City from 2016 to 2017, which can provide additional perspectives e.g. __average housing price__ on factors related with the price gap of Airbnb listings throughout NYC.

To limit the bias of our analysis, we will explore the performance of __average prices__ and __median prices__ of Airbnb listings in different neighbourhoods in NYC seperately and make some comparisons. 

Furthermore, we will scrape `Median Household Income by neighbourhood` data from `Statistical Atlas` website which presents __median household income__ in 50 most populous neighbourhoods in NYC, and then merge it to our existing dataset. The visualization and analysis of median household income can further supplement our answers to the main question of this paper -- __What leads to different Airbnb listings price in NYC?__

In the last section of this research, we will first build __linear regression models__ to __predict quantitative relationships__ between our Y variable price and some X variables we discussed in previous sections. Additionally, we will run a __regression tree__ from machine learning algorithms based on the same variables in the linear regression model. By comparing results from regression model and regression tree, we would know the advantages and drawbacks for each method, and obtain a more accurate prediction at the end.

### Maps

Now we are going to create some maps based on the NYC shape file.

### Merge New Datasets

In this section we are going to focus on more detailed location of listings (i.e. by neighbourhoods) and explore how average price of Airbnb listings change throughout the city. Additionally we will merge a new dataset containing information about housing prices on property markets in NYC from 2016 to 2017 to our original data, which can provide us more insights on explaining the price gap of Airbnb listings in NYC.

The new dataset we choose to use is the `nyc-rolling-sales.csv` from `NYC Property Sales` on Kaggle, which is a concatenated and cleaned-up version of `Rolling Sales dataset` from NYC Department of Finance.

```python
# NYC shapefile based on neighbourhoods
nyc_shape = gpd.read_file("/Users/apple/Downloads/2010 Neighborhood Tabulation Areas (NTAs)")
```

```python
# load in new dataset
housing_data = pd.read_csv("/Users/apple/Desktop/ECO225Project/Data/nyc-rolling-sales.csv")
```

```python
housing_df = pd.DataFrame(housing_data)
```

```python
#housing_df = housing_df.drop(housing_df[housing_df["SALE PRICE"] == " -  "].index)
#housing_df = housing_df[["NEIGHBORHOOD", "SALE PRICE"]]
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
# remove missing values and clean data
housing_df = housing_df.drop(housing_df[housing_df["SALE PRICE"] == " -  "].index)
housing_df = housing_df[["NEIGHBORHOOD", "SALE PRICE"]]
housing_df['exist_missing_value'] = np.where(housing_df.isnull().sum(axis=1), True, False)
clean_housing = housing_df[housing_df['exist_missing_value'] == False]
clean_housing["SALE PRICE"] = clean_housing["SALE PRICE"].astype(float)

# group new data based on neighbourhoods
gb_house = clean_housing.groupby("NEIGHBORHOOD").mean().reset_index()
gb_house = gb_house[["NEIGHBORHOOD", "SALE PRICE"]]
```

```python
# group original data by neighbourhoods
gb_neighd = clean_data1.groupby("neighbourhood")
neighd_mean = gb_neighd.mean().reset_index()
neighd_mean["neighbourhood"] = neighd_mean["neighbourhood"].str.upper()
# merge two datasets
merge_data_h = gb_house.merge(neighd_mean, left_on="NEIGHBORHOOD", right_on="neighbourhood", how="inner")
merge_data_h = merge_data_h.rename(columns = {"SALE PRICE": "average_housing_price"}).drop("NEIGHBORHOOD", axis=1)
```

```python
# merge with shapefile
nyc_shape["ntaname"] = nyc_shape["ntaname"].str.upper()
merge_data1 = nyc_shape.merge(merge_data_h, left_on="ntaname", right_on="neighbourhood", how="inner")
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
plt.figure(figsize=(35,25))
ax1 = plt.subplot(211)

nyc_shape.plot(ax=ax1, edgecolor="black", color="white")
merge_data1.plot(
    ax=ax1, edgecolor='black', column='price', legend=True, 
    cmap='Blues', alpha=0.8, legend_kwds={"label":"Average price ($)"})


plt.title("NYC Airbnb Average Price", size=15, weight="bold")
plt.axis('off')

ax2 = plt.subplot(221)
nyc_shape.plot(ax=ax2, edgecolor="black", color="white")

merge_data1.plot(
    ax=ax2, edgecolor='black', column='average_housing_price', legend=True, cmap='Blues', alpha=0.8, 
    legend_kwds={"label":"Average housing price ($)", "orientation":"vertical", 'format':"%.0f"})

plt.title("NYC Average Housing Price", size=15, weight="bold")
plt.axis('off')
```


The subplot on the right side shows the __average price level of Airbnb listings__ over all neighbourhoods in New York City. The darker the blue, the higher the average price in that area. Here we only consider neighbourhoods where the average listings price is lower than \\$220. We can see that there are 4 neighbourhoods at the __bottom right of Manhattan__ have relatively high price level on average. In fact, these neighbourhoods are close to Times Square, which is one of the NYC landmarks. The might be a reason for the high costs but we will need more evidence to support in future steps. It is worth mentioning that a neighbourhood in __Queens__ displays the darkest purple on this map, meaning the average Airbnb listings' price there is higher than any other neighbourhoods in NYC, reaching around \\$200 per night. We cannot obtain this observation if we simply look at the borough maps we created before. 

The subplot on the left side is created based on the `NYC Property Sales` dataset, it presents the __average housing prices on real estate markets__ over all neighbourhoods in NYC. We used the same geographical level in two subpots so it would be convinent to make comparison. We can observe similar trends in these 2 plots where areas with higher housing prices associates with high price level for Airbnb listings. One difference is related with 4 neighbourhoods at the __bottom right of Manhattan__ and 1 neighbourhood in __Queens__ we mentioned previously. The 4 neighbourhoods in Manhattan have the highest property prices on average. The neighbourhood in Queens with highest Airbnb price level does not have the most expensive houses on the property market on average, but its housing price is still quite high compared with other neighbourhoods. 

The similar trends between these 2 plots are reasonable. Airbnb listings are houses purchased by hosts from housing markets. If a host spends a lot on purchasing the house, he/she will charge higher prices matched with the house when the host wants to rent it on Airbnb website to prevent loss.

### Scraping New Data

In this paper our main goal is exploring possible factors affecting the Airbnb listings prices in different areas in New York City. In addition to what we already found (location, room type, annual availability and housing price on real estate market etc.), the __traffic flow__ in one area might also influence the price of listings nearby since high traffic flow is a possible indicator of high accessibility of the listing. 

In general, people would like to live in an area with convenient public transits, leading to high demand for listings with high accessibility which further increases the price for those listings. 

Therefore, we would like to consider adding data about the average traffic volume in each neighbourhood in New York City. This data can allow us to approximate the accessibility of Airbnb listings in our original dataset `AB_NYC_2019.csv`, and explore the relationship between accessibility of a listing and its price. 

The `Traffic Data Viewer - NYSDOT` is an official website provided by New York State government, displaying an interactive map that allows useer to extract traffic data information. It shows volume statistics of traffic data including the __Annual Average Daily Traffic__ for bus stations in NYC. We can use this variable to further calculate what we want -- the average traffic volume in each neighbourhood in NYC. 

We can scrape data from the `Traffic Data Viewer` website and here is the URL: `https://gisportalny.dot.ny.gov/portalny/apps/webappviewer/index.html?id=28537cbc8b5941e19cf8e959b16797b4`.

After scraping the data, we can merge it with our old dataset by neighbourhoods (either according to neighbourhood names or neighbourhood codes). We can then plotting maps to show annual average daily traffic volume for each neighbourhood and comparing it with the `NYC Airbnb Average Price` map (outcome) as well as the `NYC Average Housing Price` map (one independent variable). Additionally, we can also do scatter plots to show whether there is a linear relationship between average price of listings in one neighbourhood and the corresponding average traffic volume. This would provide some new perspectives on answering our main message __What leads to the gap of Airbnb listings in different areas in NYC?__

Since the traffic data would be updated, we should run the scraping process over time to obtain lastest data to conduct a more accurate analysis. For example, here we are interested in the annual average daily traffic thus scraping data every year would be a suitable choice. 

One thing need to mention is that since the `Traffic Data Viewer` is an interative map containing several layers and numerous statistics, it is too complex to obtain desirable data through simple scraping. We might need some external tools for web crawling such as `Scrapy` to complete the scraping process in the future.

At this stage we are going to scrape data from another website `Statistical Atlas` which containing informaiton about __Household Income in NYC__. Here is the URL: `https://statisticalatlas.com/place/New-York/New-York/Household-Income#figure/neighborhood/median-household-income`.

The consumption level of residents in the area where a listing locates might also affect how the host set price for his/her house. In general, the price level in one area is matched with people's comsumption level, and one's comsumption level relates to his/her disposable income. Households with higher disposable income would be more likely to spend money and they can afford their consumptions. We can observe in our life that goods and services are usually more expensive in the region where households have higher income. 

Here we are interested in __median household income__ by neighbourhood. The website provides the median household annual income for the 50 most populous neighbourhoods in NYC by a bar plot, and we would like to scrape this data.

```python
# import required libraries
import requests
from bs4 import BeautifulSoup
```

```python
# request the URL
web_url = 'https://statisticalatlas.com/place/New-York/New-York/Household-Income#figure/neighborhood/median-household-income'
response = requests.get(web_url)
```

```python
# obtain the soup object to see structured website contents
soup_object = BeautifulSoup(response.content, 'html.parser')
```

```python
# find the relevant tags and the graph we would like to scrape
data_table = soup_object.find_all('div', 'figure-container')[11]
```

```python
# find the values in the graph, the values are enclosed by <title>
all_values = data_table.find_all('title')
```

```python
# let the data display in a managable format 
nei_list = all_values[:53]
income_list = all_values[53:]
```

```python
# continue the cleaning process
i = 0
li1 = []
li2 = []
for nei in nei_list:
    li1.append([nei])
    
a = 0
while a < len(income_list):
    li2.append([income_list[a], income_list[a+1], income_list[a+2]])
    a += 3
    
for item in li1:
    item.extend(li2[i])
    i += 1
```

```python
# use a for loop to create a DataFrame to show values
median_household_income = pd.DataFrame(columns = ['neighbourhood', 'median_income'])
ix = 0
for row in li1:
    name = row[0].text.replace('Neighborhood', '').strip()
    median_income = row[1].text.strip('$')
    median_household_income.loc[ix] = [name, median_income] # Store it in the dataframe as a row
    ix += 1
   
    
median_household_income['median_income'] = median_household_income['median_income'].str.replace(',', '').astype(float)
median_household_income.head(5)
```

The table above shows the first 5 lines of the data we scraped from the website.

### OLS Regressions

In this section, we would like to build several linear regression models to see how each potential explanatory variable affect Airbnb listings price quantitatively. Furthermore, we can predict listings price based on their characteristics through the model we build.

At the beginning, we will determine potential variables according to background research. A paper named "Analysis of price determinants in the case of Airbnb listings" mentioned 5 groups of variables would affect Airbnb listings price: Listing Attributes, Listing Location, Listing Reputation, Host Attributes and Airbnb Policies. Under each category, there are many specific variables. For example, room type can be considered as one of listing's attributes (Toader et al., 2021). We would like to find one variable under each group and put them into our models.

First we consider variables in the original dataset `AB_NYC_2019.csv`.
For numerical variables, we would like to consider __number of reviews__ (Listing Reputation). Additionally, we want to add __annual availability__ into our model even though it does not belong to the above categories. The reason why we believe there exists a linear relationship between our outcome __Airbnb Listings Price__ and these variables comes from the scatter plot as well as maps we created in previous parts. In section Project 1 we plotted the relationship between price and number of reviews (`Price of listing V.S. Number of reviews for the listing (2019)`). From the plot we can see a negative linear trend between these 2 variables. Also in section Project 2 we created __maps__ to show average listings price among 5 boroughs in NYC as well as average annual availability of listings among 5 boroughs. Comparing 2 maps we observed a similar trend. Thus we would like to say there exists some positive linear relationship between price and annual availability, and construct a linear regression model to verify our assumptions.

For categorical variables, we would like to consider __room type__ (Listing Attributes) and __neighbourhood group__ (Listing Location). From the main plot in section Project 2, we can see observable differences of prices among different boroughs as well as among different types of room. Therefore, we will add these 2 categorical variables into our model to find out the price gap quantitatively.


The functions come from a package called `statsmodels`.

```python
# import required packages
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col
from statsmodels.formula.api import ols
```

```python
# load data
data = pd.read_csv("/Users/apple/Desktop/ECO225Project/Data/AB_NYC_2019.csv")
df = pd.DataFrame(data)
```

```python
# clean data
df['exist_missing_value'] = np.where(df.isnull().sum(axis=1), True, False)
clean_data3 = df[df['exist_missing_value'] == False]
clean_data3 = clean_data3[["price", "number_of_reviews", "availability_365", "room_type", "neighbourhood_group", "neighbourhood"]]
```

Since we want to construct linear regression models, we need to check model diagnostics, i.e. whether model assumptions are satisfied. One assumption is __normality__ of Y variable, i.e. the Y variable of linear regression should follow a normal distribution. However, based on the histogram of listings price we created in section Project 1 (`Price Distribution for New York City Airbnb Listings in 2019`) we can see that the distribution is highly skewed. In order to avoid violating the normality assumption, we would like to take __log transformation__ on our Y variable.

The histogram below shows the distribution of __log price__. The histogram is bell-shaped now, meaning log price follows a normal distribution.

```python
# data transformation
# remove 0 prices
clean_data3 = clean_data3[clean_data3["price"] > 0]
# log transformation on Y
clean_data3["log_price"] = np.log(clean_data3["price"])
```

```python

plt.hist(x="log_price", bins=20, data=clean_data3)
plt.title("Distribution of Logarithm of Price for New York City Airbnb Listings in 2019", size=15, weight='bold')
plt.xlabel("Log Price", size=13)
plt.ylabel("Density", size=13)
```

We will first build a basic model containing only numerical variables in original data. The formula would be:
$$
\log(\text{price}) = \beta_0 + \beta_1 \text{annual availability} + \beta_2 \text{number of reviews} + \epsilon
$$

Then we will add categorical variables into our model to see whether the new model has a better performance than the basic model. The formula for second regression would be:
$$
\log(\text{price}) = \beta_0 + \beta_1 \text{annual availability} + \beta_2 \text{number of reviews} + \beta_3 \text{room type} + \beta_4 \text{neighbourhood group} + \epsilon
$$

$ \beta_i $ are parameters, also called coefficients.

$ \epsilon $ is the error term, which captures randomness that cannot explained by our model.

Since room type and neighbourhood group are categorical variables, we need to convert them into __dummy variables__. Room type has 3 different categories, thus we need 2 dummies, entire apt/home is considered as reference group. Neighbourhood group has 5 categories so that we need 4 dummies where Bronx is considered as the reference group.

```python
regdata1 = pd.get_dummies(clean_data3, columns=['room_type', 'neighbourhood_group'], drop_first = True)
```

```python
# Add constant term to dataset
regdata1['const'] = 1

# Create lists of variables to be used in each regression
X1 = ['const', 'availability_365', "number_of_reviews"]
X2 = ['const', 'availability_365', "number_of_reviews",
      "room_type_Private room", "room_type_Shared room", 
     "neighbourhood_group_Brooklyn", "neighbourhood_group_Manhattan", "neighbourhood_group_Queens", 
     "neighbourhood_group_Staten Island"]

# Estimate an OLS regression for each set of variables
reg1 = sm.OLS(regdata1['log_price'], regdata1[X1], missing='drop').fit()
reg2 = sm.OLS(regdata1['log_price'], regdata1[X2], missing='drop').fit()
```

As Toader and other authors discussed in their research paper, neighbourhood average rental price may also have some impact on Airbnb listings price (Toader et al., 2021). In section Project 2 we merged a new dataset `nyc-rolling-sales.csv` containing sale prices of properties on real estate market in NYC. We would like to compute __average housing price__ (unit: dollar per square feet) in each neighbourhood and use it as an approximation of neighbourhood average rental price.
Here is the formula of the third model:
$$
\log(\text{price}) = \beta_0 + \beta_1 \text{annual availability} + \beta_2 \text{number of reviews} + \beta_3 \text{average housing price} + \beta_4 \text{room type} + \beta_5 \text{neighbourhood group} + \epsilon
$$

As we talked in section Project 2, it is reasonable that neighbourhood average housing/ rental price would affect Airbnb listings price in that neighbourhood. Airbnb is a newly developed form of rental market. The price of Airbnb listings should align with the general situation  of rental market for legal and success comptetion. For instance, if the price of Airbnb listing is much higher than the neighbourhood average rental price, then it will lose market since people will choose rental houses/ hotels with fair prices. If the price of Airbnb listing is much lower than the neighbourhood average rental price, the host may lose money and the competition may be considered as illicit. Additionally, the average housing price can reflect the amount the host paid to purchase the house (listing) to some extent. If the host paid a high price for the house, he/ she would like to set a relatively high rental price as an Airbnb listing to earn profits. We contain both neighbourhood average housing price and neighbourhood groups in our model since each neighbourhood group is quite large and includes many different neighbourhoods. The variable neighbourhood average housing price can help us explore trends within each borough.

```python

regdata_h = clean_data3
# read in housing price
housing_data = pd.read_csv("/Users/apple/Desktop/ECO225Project/Data/nyc-rolling-sales.csv")
housing_df = pd.DataFrame(housing_data)
# data cleaning
housing_df = housing_df.drop(housing_df[housing_df["SALE PRICE"] == " -  "].index)
housing_df = housing_df.drop(housing_df[housing_df["GROSS SQUARE FEET"] == " -  "].index)
housing_df = housing_df[["NEIGHBORHOOD", "SALE PRICE", "GROSS SQUARE FEET"]]
housing_df['exist_missing_value'] = np.where(housing_df.isnull().sum(axis=1), True, False)
clean_housing = housing_df[housing_df['exist_missing_value'] == False]
clean_housing["SALE PRICE"] = clean_housing["SALE PRICE"].astype(float).apply(np.int64)
# clean_housing["SALE PRICE"] = clean_housing["SALE PRICE"].apply(lambda x: format(x, 'f'))
# clean_housing["SALE PRICE"] = clean_housing["SALE PRICE"].astype(float)
clean_housing["GROSS SQUARE FEET"] = clean_housing["GROSS SQUARE FEET"].astype(float).apply(np.int64)
# clean_housing["GROSS SQUARE FEET"] = clean_housing["GROSS SQUARE FEET"].astype(float)


clean_housing["price_per_square"] = clean_housing["SALE PRICE"] / clean_housing["GROSS SQUARE FEET"]
clean_housing["price_per_square"] = clean_housing["price_per_square"].dropna().round(3)

# group new data based on neighbourhoods
gb_house = clean_housing.groupby("NEIGHBORHOOD").mean().reset_index()
gb_house = gb_house[["NEIGHBORHOOD", "price_per_square"]].dropna()
# merge with old dataset
regdata_h["neighbourhood"] = regdata_h["neighbourhood"].str.upper()
# merge two datasets
regdatah = gb_house.merge(regdata_h, left_on="NEIGHBORHOOD", right_on="neighbourhood", how="inner")
regdatah = regdatah.rename(columns = {"price_per_square": "average_housing_price"}).drop("NEIGHBORHOOD", axis=1)
regdatah['average_housing_price'] = regdatah['average_housing_price'].apply(np.int64)

# remove 0 values
regdatah = regdatah[regdatah['average_housing_price'] > 0]
min_hp = regdatah['average_housing_price'].min()

# # take transformation on average housing price

regdatah['average_housing_price'] = regdatah['average_housing_price'] - min_hp
regdatah.head(3)
```

The table above shows the first 3 lines of the merged data. Notice here we subtract each value in average housing price column by the minimum value of this column. After this transformation, a 0 average housing price is meaningful. It means the average housing price in this neighbourhood is the minimum value among all neighbourhoods in NYC.

```python
regdata6 = pd.get_dummies(regdatah, columns=['room_type', 'neighbourhood_group'], drop_first = True)
regdata6['const'] = 1
X3 = ['const', 'availability_365', "number_of_reviews", "average_housing_price",
      "room_type_Private room", "room_type_Shared room", 
     "neighbourhood_group_Manhattan", "neighbourhood_group_Queens", 
     "neighbourhood_group_Staten Island"]

reg3 = sm.OLS(regdata6['log_price'], regdata6[X3], missing='drop').fit()
```

```python
from statsmodels.iolib.summary2 import summary_col
info_dict={'R-squared' : lambda x: f"{x.rsquared:.2f}", 
           'No. observations' : lambda x: f"{int(x.nobs):d}"}
```

```python
results_table = summary_col(results=[reg1,reg2, reg3],
                            float_format='%0.8f',
                            stars = True,
                            model_names=['Model 1',
                                         'Model 2',
                                        'Model 3'],
                            info_dict=info_dict,
                            regressor_order=['const',
                                            
                                             'availability_365',
                                             'number_of_reviews',
                                              'average_housing_price',
                                            "room_type_Private room",
                                            "room_type_Shared room",
                                            "neighbourhood_group_Brooklyn",
                                             "neighbourhood_group_Manhattan", 
                                             "neighbourhood_group_Queens"])

results_table.add_title('OLS Regressions: Comparison among 3 Models')

print(results_table)
```

The table above shows the results of each regression model (estimated values of coefficients as well as corresponding p values) and indicator of model performance. The dataset we used to build model 3 has no information about average housing price (dollar per square feet) in neighbourhoods in Brooklyn, thus it is not included in model3.

Variables with stars * are considered to have significant impact on log price. R-squared and adjusted R-squared can be used to evaluate the performance of the model. R-squared, also called coefficient of determination, measures the proportion of variation of outcome that are explained by the model. The higher the R-squared, the more variation the model can explain. One drawback of R-squared is that its value is affected by number of predictors. More predictors (regardless of significance) will lead to higher R-squared. To adjust for this drawback, we instead using __adjusted R-squared__ to make comparisons between models with different sizes. 

By comparing adjusted R-squared between model 1 and model 2, we can observe that adding categorical variables into the model increases the adjusted R-squared form 0.008 to 0.495, meaning __model 2 performs much better than model 1__. 

Now we are going to see how does the performance of model changes after we adding the average housing price into the model. First we can see variable average housing price is attached with 3 stars *** , meaning it has significant impact on log price. Then we compare adjusted R-squared between 2 models. Adding this variable increases adjusted R-squared from 0.495 to 0.557. Thus we would like to conclude that __model 3 performs better than model 2 in explaining variation of outcome__.

We have some insights on how regression works on the whole dataset, we would like to dig deeper into the price gap between listings in __Manhattan__ and remaining 4 boroughs since we found that Manhattan has some distinct characteristics from our previous exploratory data analysis. Additionally, from bar plots and boxplots we created in section Project 1 we found that __entire apt/home__ listings also have highest maximum price, greatest total numbers and widest price range compared with other 2 room types. Therefore, we would like to build another model to predict the differences. 

First we do some data manipulation. We change the neighbourhood group of a listing to 1 if it locates in Manhattan and 0 otherwise. Also we change the room type of a listing to 1 if it is an entire apt/home and 0 otherwise. 

```python
# create another model
cleanup_nums = {"room_type":     {"Entire home/apt": 1, "Private room": 0, "Shared room": 0},
                "neighbourhood_group": {"Manhattan": 1, "Queens": 0, "Brooklyn": 0, "Staten Island": 0,
                                  "Bronx": 0}}

regdata5 = regdatah
regdata5 = regdata5.replace(cleanup_nums)
regdata5.head(3)
```

The table above shows the first 3 lines of the resulted data. The Airbnb listing on the first line is an entire apt/home that not in Manhattan. 

Now we are going to construct the new model. We consider __average housing price__, __annual availability__, __number of reviews__, __entire home/apt or not__ and __location in Manhattan or not__ as our potential explanatory variables. 

```python
reg5 = ols("log_price ~ average_housing_price + availability_365 + number_of_reviews + C(room_type) + C(neighbourhood_group)", data = regdata5).fit()
```

```python
# final model: reg5

plt.rc('figure', figsize=(12, 7))
plt.text(0.01, 0.05, str(reg5.summary()), {'fontsize': 10}, fontproperties = 'monospace')
plt.axis('off')
plt.tight_layout()
```

The table above shows the result summary of model 4. P values for all variables are extremely small, meaning they have significant impact on our outcome: log price. The adjusted R-squared for model 4 is 0.546, which is smaller than 0.557 from model 3 but not too much. AIC is also a tool that can be used to evaluate the balance between the goodness of fit of a model and the complexity of the model. Model with smaller AIC would be preferred. The AIC for model 3 is 1.17e+04, and the AIC for model 4 is 1.192e+04. Two models have similar AIC so that we will not criticise AIC value too much when we choose the final model. 

Now considering our context and research question. In this research we are interested in what leads to the price gap between listings in different regions in NYC. From our previous exploratory data analysis, price of listings in Manhattan has observable differences than other boroughs while the remaining 4 boroughs have similar performances. Similarly, there exists obvious gaps between price of entire apt/home listings and other types of listings. Therefore, we would like to choose __model 4 as our final model__ since it fits the context better, and has fair performance at the same time.

Below are the interpretations of the result of the final model:
- The intercept $ \hat{\beta}_0 = 3.907 $, meaning the predicted value of log price (also the expected value of log price) is 3.907 for an Airbnb listing with 0 reviews, opens for 0 days during a year, locates in a neighbourhood in boroughs other than Manhattan where the average housing price is 79 dollars per square feet (the minimum value), and whose room type is shared room/ private room.
- $ \hat{\beta}_1 = 0.7355 $ means that all other factors remain the same, the predicted log price for an entire apt/home listing is 0.7355 higher than listings with other room types.
- $ \hat{\beta}_2 = 0.3647 $ means that all other factors remain the same, the predicted log price for listings in Manhattan is 0.3647 higher than listings in other boroughs. 
- $ \hat{\beta}_3 = 0.0002 $ means that all other factors remain the same, increasing the neighbourhood average housing price by 1 dollar per square feet would increase the predicted log price of listings in that neighbourhood by 0.0002. Another way to interpret the log value is that increasing the neighbourhood average housing price by 1 dollar per square feet would increase the average listings price in that neighbourhood by 0.02 \%.
- $ \hat{\beta}_4 = 0.0006 $ means that all other factors remain the same, increasing the annual availability by 1 day would increase the predicted listing price by 0.06 \%.
- $ \hat{\beta}_5 = -0.0004 $ means that all other factors remain the same, increase the number of reviews the listing received would decrease the predicted listing price by 0.04 \%.


Here is the formula of our regression:

$$
\widehat{\log(\text{price})} = 3.907 + 0.7355* X_{\text{room type = entire apt/home}} + 0.3647* X_{\text{neighbourhood group = Manhattan}} + 0.0002* X_{\text{average housing price}} + 0.0006* X_{\text{annual availability}} - 0.0004* X_{\text{number of reviews}}
$$

The adjusted R-squared for our final model is 0.546, meaning the model can explain 54.6 \% variations of the log price.

With this model, we can predict the price of a listing based on its features (location, reputation, attributes etc.). Additionally, we can predict the price gap of listings with same characteristics but locates in different boroughs, or with different properties but locate in same boroughs. Plots and maps can give us direct visualizations about the price differences of listings, while the regression model can give us quantitative insights to help us better answer the research question -- __what leads to the price gap of listings in NYC?__

```python
corr = regdata5[["average_housing_price", "number_of_reviews", "availability_365"]].corr(method='pearson')
plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True)
plt.title("Correlation Between Each Pair of X Variables", size=15, weight='bold')
plt.xlabel("Variable name")
plt.ylabel("Variable name")
```

There is a multicollinearity warning for the final model. However, by checking the heat map which displays the correlation between each pair of X variables we do not observe high correlations. This might be one limitation of our model and need more improvement in the future.

### Machine Learning: Regression Tree

In this section we would like to apply machine learning algorithm to further improve the accuracy of our results. More specifically, we will build a regression tree and compare its result with our linear regression model.

```python
# load packages
from sklearn import tree
from sklearn import metrics
```

```python
tree_data = regdata5

y = tree_data["log_price"]

tree_data = tree_data.drop(columns = ["neighbourhood", "price"])

X = tree_data.drop(columns = ["log_price"]).copy()
# convert everything to be a float for later on
for col in list(X):
    X[col] = X[col].astype(float)

X.head()
```

```python
sqft_tree = tree.DecisionTreeRegressor(max_depth=4).fit(X,y)
```

```python
# use the fitted tree to predict
y_pred_tree = sqft_tree.predict(X)
```

*__Objective Function__*:

The objective funciton of a regression tree is to minimize error of predictions at each split. During each step the regression tree undertakes binary split and the threshold of the split is crucial and should be chosen to minimize the mean square error (MSE) of prediction. We will take one split from the tree below as an example to explain how objective function works.

Since the first split of root is based on room types which is a categorical variable with only 2 groups, the threshold value is not quite meaningful. We will use the right internal node at the second level of the whole tree, which represents entire apt/ home listings in NYC, as an example. This leaf will be splitted on average housing price. We have 2 boxes R1 and R2, where


$ R1 = X|X_{average\_housing\_price} \leq s $ and 
$ R2 = X|X_{average\_housing\_price} > s $

s is a threshold.

Our goal is to select s to minimize 
$$
\left[  \sum_{i: x_i \in R1}
    (y_i - \hat{y}_{R1})^2 +  \sum_{i: x_i \in R2}
    (y_i - \hat{y}_{R2})^2 \right]
$$
which is the sum of error of predictions in both sides of the split.

s for this split is 526 in our example.

We will continue this process to split on all variables until reaching the maximum depth of the tree we set, which is 4. The objective function for each split is similar to the example above. 

*__Regularization Parameters__*:

The max depth we set when building the tree is one of regularization parameters. It prevents the tree growing too large and too complicated. By setting the maximum depth of the tree to be 4 in our case, there will only be 4 levels under the root. If we decrease the value of regularization parameters e.g. max depth, the tree would be less complex but the mean square error might be large. If we increase the value of the regularization parameters, the tree would be much more complicated and overfitting even though the MSE is quite small. There exists a trade-off between the complexity and error of prediction of the regression tree. 





```python
sqrf_fig = plt.figure(figsize=(30,25))
sqrf_fig = tree.plot_tree(sqft_tree, feature_names=X.columns, filled=True, fontsize=10)
```

The above diagram shows the results of the regression tree. 
There are 16 branches in total and depth of the tree is 4. The terminal node for each branch shows the prediction of the regression. Here is the interpretation of one branch:

For an entire apt/home listing with neighbourhood average housing price greater than 526 dollars per square feet, open for less than 86.5 days for reservation during a year and locates in a borough other than Manhattan, the predicted value of log price is 4.889. The error of this prediction is 0.106.

```python
# find the error of prediction (MSE)
print('Mean Squared Error:', metrics.mean_squared_error(y, y_pred_tree))
```

The MSE of this regression tree is around 0.204. The error is relatively large. We can reduce the error by having a larger tree, but it is hard to visualize such a big tree and the model would be too complex to interpret. 

Now we would like to comparing the linear regression model and the regerssion tree. Due to the limit of the maximum depth, the tree does not include every X variable in each branch. This may lead to less accurate results. 

In regression tree, we split each variable into 2 groups by a threshold, while in linear regression models, we use exact values of variables to make prediction. The results of the regression tree is more understandable and easier to interpret.
