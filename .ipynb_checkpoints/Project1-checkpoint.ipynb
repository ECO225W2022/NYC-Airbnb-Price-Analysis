## Data Wrangling and Cleaning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Load in the dataset first and we will have a brief overview of the raw data.

```python
data = pd.read_csv("/Users/apple/Desktop/ECO225Project/Data/AB_NYC_2019.csv")
```

```python
df = pd.DataFrame(data)
```

There are 48895 rows and 16 columns in total.
Now we will create a new DataFrame `clean_data` by removing missing values from the original dataset and only keep the dependent variable __price__ and 3 explanatory variables: __number of reviews__, __room type__ and __neighbourhood group__.

__price__ shows the cost ($) of living one night in one particular Airbnb listing in NYC in 2019.

__number of reviews__ indicates total number of reviews one particular listing received.

__room type__ represents the listing space type. There are 3 different types: 'Private room', 'Entire home/apartment', 'Shared room'.

__neighbourhood group__ displays the location of the listing. There are 5 categories in total: 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island', 'Bronx'.


```python
df['exist_missing_value'] = np.where(df.isnull().sum(axis=1), True, False)
```

```python
clean_data = df[df['exist_missing_value'] == False]
```

```python
clean_data = clean_data[['neighbourhood_group', 'room_type', 'number_of_reviews', 'price']]
```

Below are a bunch of overview of the cleaned data we will use to conduct analysis. The Y variable __price__ and explanatory variable __number of reviews__ are integers; other 2 explanatory variables __neighbourhood group__ and __room type__ are categorical variables, having 5 and 3 categories respectively.

## Exploratory Data Analysis (Summary Statistics)

Starting with understanding and exploring the data, we will first look into features of the outcome we are interested -- __price__ of listings.

The histogram below provides an intuitive insight of the distribution of prices of Airbnb listings in NYC in 2019.

Most of the listings priced below \\$1000 with some extreme values reaching \\$10000 dollars per night.


```python
sns.histplot(data=clean_data, x='price', binwidth=30)
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style({'axes.grid' : False})
plt.title("Price Distribution for New York City Airbnb Listings", size=15, weight='bold')
plt.xlabel("Price", size=13)
plt.ylabel("Number of listings", size=13)
```

For better visualization we will remove the outliers and draw the histogram again.

```python
sub_price = clean_data[clean_data['price'] < 500]
```

```python
sns.histplot(data=sub_price, x='price', binwidth=30)
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style({'axes.grid' : False})
plt.title("Price Distribution for New York City Airbnb Listings (exclude extreme values)", size=15, weight='bold')
plt.xlabel("Price", size=13)
plt.ylabel("Number of listings", size=13)
```

Now the histogram look betters and shows a clear trend of the listings' prices. We can see from the diagram that most listings' cost per night stays within \\$30 - \\$ 200. The price distribution is right skewed with the number of listings whose price set between \\$60 and \\$100 exceeds 8000, which is much greater than other price ranges. 

Below is the histogram for number of reviews that NYC Airbnb listings received in 2019. 
It is extremely right skewed and there exists outliers we need to deal with.

```python
sns.histplot(data=clean_data, x='number_of_reviews', binwidth=3)
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style({'axes.grid' : False})
plt.title("Number of Reviews for New York City Airbnb Listings", size=15, weight='bold')
plt.xlabel("Number of reviews", size=13)
plt.ylabel("Number of listings", size=13)
```

Similarly, we would like to remove extreme values for better visualization.

Now the histogram looks better and as shown on the diagram, more than 10000 listings received around 1 - 3 reviews in total. Only 5000 listings had 3-6 reviews approximately.  The histogram displays an obvious decreasing trend on number of listings when number of reviews increases.

```python
sub_review = clean_data[clean_data['number_of_reviews'] < 100]
```

```python
sns.histplot(data=sub_review, x='number_of_reviews', binwidth=3)
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style({'axes.grid' : False})
plt.title("Number of Reviews for New York City Airbnb Listings (exclude extreme values)", size=15, weight='bold')
plt.xlabel("Number of reviews", size=13)
plt.ylabel("Number of listings", size=13)
```

Now we would like to look at some numerical summaries for __number of reviews__ (X variable) and __price__ (Y variable).

The average price for NYC airbnb is around \\$142 per night in 2019, while the most expensive room reaches \\$10000 per night. The interquantile range for __price__ is 170 - 69 = \\$101, measuring the spread of the middle 50 \% of the data. That is, the middle 50 \% of the price is between \\$69 and \\$170. 

Each NYC airbnb listing received around 29 reviews on average in 2019. The most commented listing had 629 reviews and the least commented listing only had 1 review. Number of reviews that middle 50 \% of listings received ranges from 3 to 33.

```python
clean_data.describe()
```

The scatterplot below displays the relationship between price of listings and the number of reviews the listing received. The plot shows a __negative__ trend where listings with lower cost obtained more reviews. One possible reason behind this trend might be that more people may choose to book rooms with lower price, thus more people had experience with those cheaper rooms and may leave their comments. 

```python
sns.scatterplot(x='number_of_reviews', y='price', data=clean_data)
sns.set(rc={'figure.figsize':(20,8)})
plt.title("Price of listing V.S. Number of reviews for the listing", size=15, weight='bold')
plt.xlabel("Number of reviews", size=13)
plt.ylabel("Price", size=13
```

After understanding the distribution of price and number of reviews as well as their relationships, we are now going to consider the 2 categorical variables __neighbourhood groups__ and __room type__.

First we will look at the distribution of listings among 5 neighbourhood groups.

```python
                 
sns.countplot(x='neighbourhood_group', data=clean_data)
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style({'axes.grid' : False})
plt.title("Number of listings in 5 Neighbourhood groups", size=15, weight='bold')
plt.xlabel("Neighbourhood group", size=13)
plt.ylabel("Number of listings", size=13)
```

The bar plot above shows number of listings in 5 neighbourhood groups in NYC in 2019. Manhattan and Brooklyn have top 2 number of listings (each with more than 16000), almost 3 times number of listings in Queens. Much fewer listings (below 1000) locate in Staten Island as well as Bronx compared with Queens region. 

```python

sns.countplot(x='room_type', data=clean_data)
sns.set(rc={'figure.figsize':(12,8)})
sns.set_style({'axes.grid' : False})
plt.title("Number of listings with different room types", size=15, weight='bold')
plt.xlabel("Room type", size=13)
plt.ylabel("Number of listings", size=13)
```

Now we would like to understand how listings are distributed among different room types in NYC. 

As shown in the bar plot above, near 20000 listings in NYC are Entire home/apt, which is the most popular room type. Private room ranks as the second popular room type, contributing around 17500 listings on Airbnb website while number of Shared room is the lowest.

After understanding the dsitribution of listings among neighbourhood groups and different room types seperately, we  would like to combine them together to explore how many listings with different room types are distributed over the 5 neighbourhood group. From the table below we can see that Brooklyn and Manhattan are 2 regions provide the greatest amount of Entire home/apt and Private room.
In Brooklyn, there are around 8000 Entire home/apt listed on Airbnb website (same for Private room) while only 290 Shared room. Similar situation happens in Manhattan. 9962 Entire home/apt and 6303 Private room are available on Airbnb website while there exists only 356 Shared room.

Compared with Brooklyn and Manhattan, fewer listings are located in Queens and number of listings plunge in Bronx and Staten Island despite the type of room. In these 3 regions, Private room is more popular than other types of room.

Since we consider __neighbourhood group__ as one of the factors affecting __price__ of listings, we would like to examine relationship between these 2 variables.

```python
sns.catplot(x='neighbourhood_group', y='price',kind='box',height=6, aspect=2, data=sub_price)
plt.xlabel("Neighbourhood group", size=13)
plt.ylabel("Price", size=13)
plt.title("Distribution of listings' prices in 5 neighbourhood groups", weight='bold', size=15)
```

The diagram above displays distribution of prices of listings in each neighbourhood group separately via 5 boxplots. Dots lie outside the whiskers of boxplots are considered as outliers. 

As we can see above, listings in Queens, Staten Island and Bronx have similar price range (exclude outliers), going from \\$0 to \\$200. Rooms in Manhattan have much wider price range (exclude outliers) compared with other neighbourhoods, going from \\$1 to \\$360 approximately. 

In general, it would cost more if we choose to live in Manhattan and Brooklyn is the second expensive.

In addition, we also take interest in the relationship between __price__ of listings and their __room type__. Same as before, we will use boxplots to visualize this association.

```python
sns.catplot(x='room_type', y='price',kind='box',height=6, aspect=2, data=sub_price)
plt.xlabel("Room type", size=13)
plt.ylabel("Price", size=13)
plt.title("Distribution of listings' prices among different room types", weight='bold', size=15)
```

As shown above, Entire home/apt has the widest price range, going from \\$0 to approximate \\$320. Shared room has the most narrowed range of price, going from \\$0 to approximate \\$120. Private room is in between, going from \\$0 to approximate \\$150. Also Entire home/apt has a greater interquantile range as well compared with other 2 types of room, meaning the corresponding price data have higher spread. '

One possible reason for this trend is that Entire home/apt provides more space and privacy compared with other 2 types of room, leading to a higher price in general. 



Finally, we would like to figure out the distribution of prices over more detailed subsets thus we group the data based on 2 factors: __neighbourhood group__ and __room type__ to produce the plot below.

```python
sns.catplot(x='neighbourhood_group', y='price', hue='room_type', kind='box',height=6, aspect=2, data=sub_price)
plt.xlabel("Neighbourhood group", size=13)
plt.ylabel("Price", size=13)
plt.title("Distribution of listings' prices among different room types in 5 neighbourhood groups", weight='bold', size=15)
```

As we can see, Entire home/apt in Manhattan has the widest price range from \\$1 to \\$400 approximately (exclude outliers), while Shared room in Brooklyn, Queens and Bronx have the lowest price ranges (exclude outliers) which go from \\$10 to \\$60 approximately.
